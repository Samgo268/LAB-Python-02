# Ejercicio 4 - Reduce

```python
from functools import reduce
```

## a. Multiplicación de todos los elementos de una lista

```python
numeros = [1, 2, 3, 4, 5]

producto_total = reduce(lambda a, b: a * b, numeros)
print(producto_total) 
```

## b. Concatenar una lista de palabras en una sola cadena

```python
palabras = ['Hola', ' ', 'mundo']

frase = reduce(lambda a, b: a + b, palabras)
print(frase)
```

## c. Hallar el mayor de todos usando función auxiliar

```python
def el_mayor(a, b):
    return a if a > b else b

numeros = [3, 1, 7, 2, 9, 4, 6]

maximo = reduce(el_mayor, numeros)
print(maximo) 
```

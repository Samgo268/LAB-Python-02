# Ejercicio 4 - Filter

## a. Palabras con más de 4 letras

```python
palabras = ['sol', 'montaña', 'río', 'estrella', 'mar', 'luna', 'planeta']

palabras_largas = list(filter(lambda p: len(p) > 4, palabras))
print(palabras_largas)
```

## b. Filtrar elementos None de una lista

```python
lista = [1, None, 3, None, 5, None, 7, 8]

sin_nulos = list(filter(lambda x: x is not None, lista))
print(sin_nulos)
```

## c. Palabras que empiezan por una vocal

```python
palabras = ['avión', 'barco', 'elefante', 'iguana', 'oso', 'tigre', 'uva', 'pato']
vocales = 'aeiouáéíóúAEIOUÁÉÍÓÚ'

con_vocal = list(filter(lambda p: p[0] in vocales, palabras))
print(con_vocal)
```

## d. Palabras que son palíndromos

```python
palabras = ['oso', 'amor', 'reconocer', 'python', 'anilina', 'casa', 'radar']

palindromos = list(filter(lambda p: p == p[::-1], palabras))
print(palindromos)
```

## e. Números que terminan en 5

```python
numeros = [5, 10, 15, 22, 25, 30, 35, 41, 45, 50]

terminan_en_5 = list(filter(lambda n: n % 10 == 5, numeros))
print(terminan_en_5)
```

## f. Productos con precio mayor a 100

```python
productos = [
    {'nombre': 'Teclado', 'precio': 80},
    {'nombre': 'Mouse', 'precio': 40},
    {'nombre': 'Monitor', 'precio': 300},
    {'nombre': 'Webcam', 'precio': 150}
]

productos_caros = list(filter(lambda p: p['precio'] > 100, productos))
print(productos_caros)
```

# Ejercicio 1 - Funciones Lambda
Definición y uso de funciones lambda en Python.

## a. Lambda: ¿es múltiplo de 3?

```python
es_multiplo_3 = lambda n: n % 3 == 0

print(es_multiplo_3(9))   # True
print(es_multiplo_3(10))  # False
print(es_multiplo_3(0))   # True
```

## b. Lambda: cubo de un número

```python
cubo = lambda n: n ** 3

print(cubo(3))   # 27
print(cubo(5))   # 125
print(cubo(-2))  # -8
```

## c. Lambda: producto de dos números

```python
producto = lambda a, b: a * b

print(producto(4, 5))   # 20
print(producto(3, -2))  # -6
```

## d. Lambda: mayor de dos números

```python
mayor = lambda a, b: a if a > b else b

print(mayor(7, 3))   # 7
print(mayor(2, 10))  # 10
print(mayor(5, 5))   # 5
```

## e. Lambda: ¿empieza con 'A' o 'a'?

```python
empieza_con_a = lambda palabra: palabra[0].lower() == 'a'

print(empieza_con_a('Avión'))    # True
print(empieza_con_a('árbol'))    # True
print(empieza_con_a('Casa'))     # False
```

## f. Lambda: Celsius a Fahrenheit

```python
celsius_a_fahrenheit = lambda c: c * 9/5 + 32

print(celsius_a_fahrenheit(0))    # 32.0
print(celsius_a_fahrenheit(100))  # 212.0
print(celsius_a_fahrenheit(37))   # 98.6
```

## g. Lista de tres lambdas y aplicación al número ingresado por el usuario

```python
operaciones = [
    lambda n: n * 2,       # Duplicar
    lambda n: n + 10,      # Sumar 10
    lambda n: n ** 2       # Cuadrado
]

numero = int(input('Ingrese un número: '))

for operacion in operaciones:
    print(operacion(numero))
```

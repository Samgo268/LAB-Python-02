# Ejercicio 4 - Sorted

## a. Ordenar palabras alfabéticamente

```python
palabras = ['banana', 'manzana', 'uva', 'cereza', 'durazno']

ordenadas = sorted(palabras)
print(ordenadas)
```

## b. Ordenar lista de tuplas por el primer elemento

```python
tuplas = [(3, 'c'), (1, 'z'), (2, 'b'), (5, 't'), (4, 'o')]

ordenadas = sorted(tuplas, key=lambda t: t[0])
print(ordenadas)
```

## c. Ordenar productos por precio (mayor a menor) y luego por nombre

```python
productos = [
    ('Camisa', 30),
    ('Pantalón', 55),
    ('Medias', 10),
    ('Chaqueta', 80)
]


ordenados = sorted(productos, key=lambda p: (-p[1]))
ordenadosn = sorted(productos, key=lambda p: (p[0]))
print(ordenados)
print(ordenadosn)
```

## d. Ordenar palabras según su longitud

```python
palabras = ['sol', 'montaña', 'río', 'estrella', 'mar', 'planeta']

por_longitud = sorted(palabras, key=lambda p: len(p))
print(por_longitud)
```

## e. Pares primero, luego impares

```python
numeros = [5, 2, 8, 1, 4, 7, 6, 3, 9, 10]

pares_primero = sorted(numeros, key=lambda n: n % 2)
print(pares_primero)
```

## f. Frecuencia de letras en una palabra, ordenada por frecuencia

```python
palabra = 'programacion'
frecuencias = []

def frecuencia(palabra, frecuencias):
    for letra in palabra:
        n = 0
        for i in range(len(palabra)):
            if letra == palabra[i]:
                n += 1
        frecuencias.append(n)
    return frecuencias

print(frecuencia(palabra, frecuencias))

# Ordenar por frecuencia descendente
#diccionario_ordenado = dict(sorted(frecuencias.items(), key=lambda item: -item[1]))
#print(diccionario_ordenado)
```

## g. Empleados: ordenar por departamento y luego por edad

```python
empleados = [
    ('Maria', 'Ventas', 30),
    ('Luis', 'Tecnología', 25),
    ('Ana', 'Ventas', 25),
    ('Pedro', 'Tecnología', 28)
]

ordenados = sorted(empleados, key=lambda e: (e[1], e[2]))
print(ordenados)
```

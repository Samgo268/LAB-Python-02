# Ejercicio 3 - Map

## a. Temperaturas Celsius a Fahrenheit con map

```python
celsius = [0, 20, 37, 100, -10]

fahrenheit = list(map(lambda c: c * 9/5 + 32, celsius))
print(fahrenheit)
```

## b. Sumar elementos de dos listas posición a posición

```python
lista1 = [1, 2, 3, 4, 5]
lista2 = [10, 20, 30, 40, 50]

suma_listas = list(map(lambda a, b: a + b, lista1, lista2))
print(suma_listas)
```

## c. Áreas de círculos dado sus radios

```python


radios = [1, 2, 3, 4, 5]

areas = list(map(lambda r: 3.1415 * r**2, radios))
print([round(a, 4) for a in areas])
```

## d. Diccionario de precios → lista de tuplas con precio +10%

```python
productos = {'pan': 1000, 'leche': 2500, 'café': 5000}

con_incremento = list(map(lambda item: (item[0], item[1] * 1.10), productos.items()))
print(con_incremento)
```

## e. Multiplicar todos los elementos de una matriz por 10

```python
matriz = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

matriz_por_10 = list(map(lambda fila: list(map(lambda x: x * 10, fila)), matriz))
print(matriz_por_10)
```
